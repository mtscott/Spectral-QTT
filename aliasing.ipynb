```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from functools import partial
```

```python
x = np.linspace(0, 2*np.pi, 500)
y = np.arange(0,6.1, 0.5)
f1 = np.sin(np.pi * x)
f11 = np.sin(np.pi * y)
f2 = np.sin(5*np.pi * x)
f3 = np.sin(9*np.pi * x)

plt.figure(figsize=(12,6))
plt.plot(y,f11,'o')

plt.plot(x,f2)
#plt.plot(x,f3)
plt.plot(x,f1)
plt.xlim([-0.1,6.1])
plt.ylim([-1.1,1.1])
plt.xlabel('Time') 
plt.ylabel('Signal') 
plt.title('Reconstructing Signals') 
plt.grid(True) 
  
plt.show()
```

```python
fig, ax = plt.subplots()
t = np.linspace(0.0, 6.0, 12)
z = np.sin(t)

scat = ax.scatter(t[0], z[0], c="b", s=5, label=f'observed signal')
ax.set(xlim=[-0.1, 6.1], ylim=[-1.1, 1.1], xlabel='Time [s]', ylabel='signal')
ax.legend()


def update(frame):
    # for each frame, update the data stored on each artist.
    x = t[:frame]
    y = z[:frame]
    # update the scatter plot:
    data = np.stack([x, y]).T
    scat.set_offsets(data)
    
    return (scat)


ani = FuncAnimation(fig=fig, func=update, frames=40, interval=30)

ani.save(filename="/Users/mitchellscott/Desktop/pillow.png", writer="pillow")
```
